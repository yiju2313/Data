```
install.packages ("SportsAnalytics")
install.packages ("ggplot2")

library (SportsAnalytics)
library (ggplot2)

NBA1617 <- fetch_NBAPlayerStatistics("16-17")

str(NBA1617)
```

安裝和載入SportsAnalyti和ggplot2。用str叫出NBA1617的所有數據。

```
NBA1617Order <- NBA1617[order(NBA1617$Steals, decreasing = T),]
NBA1617Order[1:30,]
```

利用order()函數，選取Steals由大到小排序，取前30個。

```
subset(NBA1617, Team == "ATL")
```

用subset搜尋ATL，可得所有有ALU的資料。

```
NBA1617[grepl("Steven", NBA1617$Name),]
```

用grepl搜尋Steven，可得Steven在Name裡面的資料。

```
qplot(GamesPlayed, TotalPoints, data = NBA1617)
```

TotalPoints的直方圖，並用Team塗顏色。

```
df <- data.frame (gp = factor(rep(letters[1:3], each = 10)), y = rnorm(30))

ggplot(df, aes(x = gp, y = y)) + geom_boxplot( color = 2)
```
